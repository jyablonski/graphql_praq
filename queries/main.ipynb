```python
import requests
import json
import pandas as pd

table = 'allTeams'
```

```python
query = """
query {
  allTeams {
    activeInjuries
    activeProtocols
    conference
    last10
    rank
    losses
    teamFull
    wins
    winPct
    team
  }
}
"""
```

```python
url = 'https://jyablonski_graphql.deta.dev/graphql'
r = requests.post(url, json={'query': query})
print(r.status_code)
# print(r.text)
```

```python
df = pd.DataFrame(r.json()['data'][table])
df['table'] = table
df
```
